```python
from ultralytics import YOLO

# Ścieżka do pliku konfiguracyjnego dataset
data_config = r'/content/drive/MyDrive/Colab_Notebooks/YOLO_znaki/dataset.yaml'

# Załadowanie pre-trenowanego modelu
model = YOLO('yolov8n.pt')

results = model.train(
    data='/content/drive/MyDrive/Colab_Notebooks/YOLO_znaki/dataset.yaml',
    epochs=100,
    imgsz=640,
    device='0',  # 'cpu' jeśli nie masz GPU
    patience=10,  # Wczesne zatrzymanie jeśli brak poprawy
    lr0=0.01,
)

# Ocena modelu na zestawie testowym
metrics = model.val()
print(metrics)
```
